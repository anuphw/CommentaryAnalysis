```python
import csv
import pandas as pd

myd = pd.read_csv('commentary.csv',sep=',',header=None)
myd.columns = ['id1','id2','homeTeam','awayTeam','date','time','minute','comment']
```

```python
def preProcess(text):
    import re
    return re.sub('[^a-z]+',' ',str(text).lower())

myd.processedComment = myd.comment.apply(preProcess)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(myd.processedComment)
```

```python
true_k = 20
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print ("Cluster %d:" % i)
    for ind in order_centroids[i, :10]:
        print (' %s' % terms[ind])
    print
```
